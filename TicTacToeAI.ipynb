```python
#Prints Board and Checks Winner
def printboard(gb):
    print(gb[0:3])
    print(gb[3:6])
    print(gb[6:])
def checkwinner(gb):
    for i in range(3):
        if (gb[i] == gb[i+3] and gb[i+3] == gb[i+6] and gb[i] != " "):
            return (True,gb[i])
        if (gb[3*i] == gb[3*i + 1] and gb[3*i + 1] == gb[3*i + 2] and gb[3 * i] != " "):
            return (True,gb[3*i])
    #Diagonals 
    if (gb[0] == gb[4] and gb[4] == gb[8] and gb[0] != " "):
        return (True,gb[0])
    if (gb[2] == gb[4] and gb[4] == gb[6] and gb[2] != " "):
        return (True,gb[2])
    #Checking Tie
    if " " not in gb:
        return (True,"T")
    return (False,"Continue")    
```

```python
#Training Game
import random 
def game():
    gameboard = [" "," "," "," "," "," "," "," "," "]
    numberstopick = [0,1,2,3,4,5,6,7,8]
    numbersselected = []
    allstates = []
    winfound = False
    winner = " "
    while True:
        xselection = random.randint(0,len(numberstopick) - 1)
        xplayerpos = numberstopick.pop(xselection)
        numbersselected.append(xplayerpos)
        gameboard[xplayerpos] = "X"
        #Checks winner after X plays
        winfound,winner = checkwinner(gameboard)
        if winfound == True:
            break
        #Selects O
        oselection = random.randint(0,len(numberstopick)-1)
        oplayerpos = numberstopick.pop(oselection)
        numbersselected.append(oplayerpos)
        gameboard[oplayerpos] = "O"
        #Checks Winner After O Plays
        winfound,winner = checkwinner(gameboard)
        if winfound == True:
            break
    numbersselected.append(winner)
    return numbersselected
```

```python
#Training Model
import pickle
xwins = 0 
owins = 0 
ties = 0
#States is loaded from a file
states = pickle.load( open( "save.p", "rb" ) )
for i in range(10000):
    stepstaken = game()
    write = ""
    for x in stepstaken[:-1]:
        write += str(x)
        if write not in states: 
            if stepstaken[-1] == 'X': 
                states[write] = 1
            if stepstaken[-1] == 'O': 
                states[write] = -1  
            if stepstaken[-1] == "T":
                states[write] = 0
        else: 
            if stepstaken[-1] == 'X': 
                states[write] += 1
            if stepstaken[-1] == 'O': 
                states[write] += -1  
            if stepstaken[-1] == "T":
                states[write] += 0
pickle.dump(states, open( "save.p", "wb" ) )
```
